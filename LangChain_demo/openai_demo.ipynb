# OpenAI 流式输出

内容一段一段输出，设置`strem=True`

```python
import os
from openai import OpenAI

client = OpenAI(
    api_key="sk-cp-cCklcsA3A7xhoFLbQvAKlZCQ3t19DKYO__BJ-QXrs3PH4ED3VELD37UG4Gl2TYShKMUVl9G5mLWb6mgKXz4Ju6ceP-6mzLP10rRCYAp7YHTUFQCqUQuELDc",
    base_url="https://api.minimaxi.com/v1"
)

response = client.chat.completions.create(
    model="MiniMax-M2.7",
    
    # arg:
    #     role:是一个字符串，表示消息的角色，可以是"system"、"user"、"assistant"或"assertion"。
    #     content:是一个字符串，表示消息的内容，可以是任何文本。
    #     messages:是一个列表，包含多个消息对象，每个对象都有role和content字段。
    
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "assistant", "content": "MiniMax-M2.7是一个强大的语言模型，能够理解和生成自然语言文本。"},
        {"role": "user", "content": "你是谁？你能做什么？"},
    ],
    stream=True,  # 开启流式输出
    
)
for  chunk in response:
    if chunk.choices[0].delta.content:
        print(chunk.choices[0].delta.content, end='', flush=True)
    
```

# OpenAI附带历史消息调用模型

调用模型传入的参数messages，其要求是list对象，即表明其支持非常多的消息在内。可以基于此，将历史消息填入，让模型知晓对话的上下文，更好的回答

```python
import os
from openai import OpenAI

client = OpenAI(
    api_key="sk-cp-cCklcsA3A7xhoFLbQvAKlZCQ3t19DKYO__BJ-QXrs3PH4ED3VELD37UG4Gl2TYShKMUVl9G5mLWb6mgKXz4Ju6ceP-6mzLP10rRCYAp7YHTUFQCqUQuELDc",
    base_url="https://api.minimaxi.com/v1"
)

response = client.chat.completions.create(
    model="MiniMax-M2.7",
    
    # arg:
    #     role:是一个字符串，表示消息的角色，可以是"system"、"user"、"assistant"或"assertion"。
    #     content:是一个字符串，表示消息的内容，可以是任何文本。
    #     messages:是一个列表，包含多个消息对象，每个对象都有role和content字段。
    
    messages=[
        {"role": "system", "content": "你是个AI助手，回答很简洁."},
        {"role": "user", "content": "23级人工智能1班有58人"},
        {"role": "assistant", "content": "好的，我知道了。"},
        {"role": "user", "content": "23级人工智能2班有57人"},
        {"role": "assistant", "content": "好的，我知道了。"},
        {"role": "user", "content": "23级人工智能有多少人？"},
    ],
    stream=True,  # 开启流式输出
    
)
for  chunk in response:
    if chunk.choices[0].delta.content:
        print(chunk.choices[0].delta.content, end='', flush=True)
    
```
